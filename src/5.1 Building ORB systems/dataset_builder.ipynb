```python
import numpy as np
import cv2
cap = cv2.VideoCapture('../vid/swarm_drone.mp4')

i = 0
while True:
    cap.set(cv2.CAP_PROP_POS_FRAMES, i)
    ret, frame = cap.read()
    if ret == False or cv2.waitKey(60) & 0xff == 27:
        break
    cv2.imwrite('../img/swarm_drone_skip/swarm_drone_skip{}.jpg'.format(i), frame)
    print(i)
    i += 100
cv2.destroyAllWindows()
cap.release()
```

```python
import os
files = [file for file in os.listdir('../img/label_drone/') if 'classes.txt' not in file]
pics = []
metadata = []
for file in files:
    if 'jpg' in file:
        pics.append(file)
    else:
        metadata.append(file)
```

```python
pics_and_meta = list(zip(pics, metadata))
```

```python
data = pics_and_meta[1]
def show(data):
    pic, meta = data
    _, frac_x, frac_y, frac_width, frac_height = parse_meta(meta)
    img = cv2.imread('../img/label_drone/' + pic)
    max_y, max_x, _ = img.shape
    x, y = np.int(max_x * frac_x), np.int(max_y * frac_y)
    width, height = np.int(max_x * frac_width / 2), np.int(max_y * frac_height / 2)
    crop = img[y - height: y + height, x - width: x + width]
    cv2.imshow('crop', crop)
    cv2.circle(img,(x, y), 5, (0, 255, 0), -1)
    cv2.rectangle(img,(x - width, y - height),(x + width, y + height),(0,255,0),3)
    cv2.imshow('img', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
show(data)
```

```python
def parse_meta(meta):
    with open('../img/label_drone/' + meta) as f:
        values = f.readline().split(' ')
        values = [float(value) for value in values]
    return values
parse_meta(pics_and_meta[0][1])
```
