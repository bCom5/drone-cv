```python
import cv2
import numpy as np
link='../../vid/drone-flight.mp4'

def scale(frame, factor=.5):
    height, wid, _ = frame.shape
    scaled_wid, scaled_height = np.int(wid * factor), np.int(height * factor)
    frame = cv2.resize(frame, (scaled_wid, scaled_height))
    return frame


cap = cv2.VideoCapture(link)
while True:
     # Capture frame-by-frame
    ret, frame = cap.read()
    frame = scale(frame, factor=.25)
    orb = cv2.ORB_create()

    # find the keypoints with ORB
    kp = orb.detect(frame, None)

    # compute the descriptors with ORB
    kp, des = orb.compute(frame, kp)
    print(des)
    frame2 = cv2.drawKeypoints(frame, kp, None, color=(0,255,0), flags=0)
    cv2.imshow('frame2', frame2)
    k = cv2.waitKey(60) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
